```python
# Replace 'your_data_path' with the actual path to your dataset file
data_path = 'USDA TN Food Access Data 2019 - Sheet1 (1).csv'
df = pd.read_csv(data_path)


```

```python

tn_df = df[df['State'] == 'Tennessee']

# Sort the Tennessee DataFrame by MedianIncome in ascending order
tn_df_sorted = tn_df.sort_values(by='MedianFamilyIncome')

# Retrieve the three counties with the lowest median income
lowest_income_counties = tn_df_sorted[['County', 'MedianFamilyIncome']].head(3)

# Display the result
print(lowest_income_counties)
```

```python
#Population count beyond 10 miles from supermarket
# Sort the Tennessee DataFrame by lapop10 in descending order (highest first)
tn_df_sorted = tn_df.sort_values(by='lapop10', ascending=False)

# Retrieve the three counties with the highest access population per 10 miles
highest_access_counties = tn_df_sorted[['County', 'lapop10']].head(3)

# Display the result
print(highest_access_counties)
```

```python
#Poverty Rate
# Sort the Tennessee DataFrame by PovertyRate in descending order (highest first)
tn_df20_sorted = tn_df.sort_values(by='PovertyRate', ascending=False)

# Retrieve the county with the highest poverty rate
highest_poverty_rate_county = tn_df_sorted[['County', 'PovertyRate']].head(5)

# Display the result
print(highest_poverty_rate_county)
```

```python
# Sort the Tennessee DataFrame by LAPOP1_20 in descending order (highest first)
tn_df_sorted = tn_df.sort_values(by='LAPOP1_20', ascending=False)

# Get the top 5 counties with the highest values
top_5_counties = tn_df_sorted[['County', 'LAPOP1_20']].head(5)

# Create a bar plot to visualize the data
plt.figure(figsize=(10, 6))
plt.barh(top_5_counties['County'], top_5_counties['LAPOP1_20'])
plt.xlabel('Low Access Population at 1 Mile (Urban) and 20 Miles (Rural)')
plt.ylabel('County')
plt.title('Top 5 Counties with Highest Low Access Population')
plt.gca().invert_yaxis()  # Invert the y-axis to show the highest values at the top
plt.tight_layout()

# Show the plot
plt.show()
```
