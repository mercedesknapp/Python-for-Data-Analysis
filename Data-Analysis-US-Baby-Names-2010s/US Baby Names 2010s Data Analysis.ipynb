```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('US Baby Names 2010.csv')
```

```python
df.head()
```

```python
type(df)
```

```python
print(df.describe())
```

```python
print(df.info())
```

```python
#Most Common Name

most_common_name = df['Names'].value_counts().idxmax()

print(f"The most common name is: {most_common_name}")
```

```python
#Most Common Male Name
male_data = df[df['Gender'] == 'M']

# Count male names
male_name_counts = male_data['Names'].value_counts()

# Find the most common male name
most_common_male_name = male_name_counts.idxmax()

print(f"The most common male name is: {most_common_male_name}")
```

```python
#summary statistics:
summary_stats = df['Number_of_babies'].describe()

print(summary_stats)
```

```python
# Visualize outliers using a box plot
plt.boxplot(df['Number_of_babies'])
plt.show()

# Define a function to remove outliers (e.g., using Z-scores)
from scipy import stats
z_scores = stats.zscore(df['Number_of_babies'])
df = df[(z_scores < 3)]

```

```python
#bar chart for Top 10 Baby Names by Number of Babies
name_counts = df.groupby('Names')['Number_of_babies'].sum().reset_index()

# Sort the data by the number of babies in descending order
name_counts = name_counts.sort_values(by='Number_of_babies', ascending=False)

# Select the top 10 names
top_10_names = name_counts.head(10)

# Create a bar chart for the top 10 names
plt.figure(figsize=(10, 6))
plt.barh(top_10_names['Names'], top_10_names['Number_of_babies'], color='skyblue')
plt.xlabel('Number of Babies')
plt.ylabel('Names')
plt.title('Top 10 Baby Names by Number of Babies')
plt.gca().invert_yaxis()
plt.show()
```

```python
name_counts = df.groupby('Names')['Number_of_babies'].sum().reset_index()

# Sort the data by the number of babies in ascending order (lowest to highest)
name_counts = name_counts.sort_values(by='Number_of_babies', ascending=True)

# Select the lowest 10 names
lowest_10_names = name_counts.head(10)

# Create a bar chart for the lowest 10 names
plt.figure(figsize=(10, 6))
plt.barh(lowest_10_names['Names'], lowest_10_names['Number_of_babies'], color='skyblue')
plt.xlabel('Number of Babies')
plt.ylabel('Names')
plt.title('Lowest 10 Baby Names by Number of Babies')
plt.show()


```
